```python
def plot_diff(model, y_mirror, error_mirror, y_no_mirror, error_no_mirror, labels):
    fig, ax = plt.subplots()
    x_pos = np.arange(len(labels))
    ax.bar(x_pos-0.2, y_mirror, 0.4, yerr=error_mirror, align='center', capsize=10, label='Mirror method')
    ax.bar(x_pos+0.2, y_no_mirror, 0.4, yerr=error_no_mirror, align='center', capsize=10, label='Basic evaluation')
    ax.set_xticks(x_pos)
    ax.set_xticklabels(labels)
    ax.yaxis.grid(True)
    ax.set_title(model)

    plt.tight_layout()
    plt.legend()
    plt.show()
```

```python
from matplotlib.transforms import Affine2D
def plot_error(model, y_mirror, error_mirror, y_no_mirror, error_no_mirror, labels):
    fig, ax = plt.subplots()
    x_pos = np.arange(len(labels))

    trans1 = Affine2D().translate(-0.1, 0.0) + ax.transData
    trans2 = Affine2D().translate(+0.1, 0.0) + ax.transData

    ax.plot(x_pos, y_mirror, label='Mirror method')
    ax.errorbar(x_pos, y_mirror, yerr = error_mirror, fmt ='o', color='blue', transform=trans1)
    ax.plot(x_pos, y_no_mirror, label='Basic evaluation')
    ax.errorbar(x_pos, y_no_mirror, yerr = error_no_mirror, fmt ='o', color='orange', transform=trans2)
    ax.set_xticks(x_pos)
    ax.set_xticklabels(labels)
    ax.yaxis.grid(True)
    ax.set_title(model)

    plt.tight_layout()
    plt.legend()
    plt.show()
```
